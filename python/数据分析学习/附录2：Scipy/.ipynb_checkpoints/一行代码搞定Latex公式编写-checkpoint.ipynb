写论文时，手敲成堆的计算公式，被虐到头秃？

做PPT时，几十页的计算推理公式，恨不得直接手写拍照？

现在，解放双手的时刻来了，只需要掌握一点Python语法，上面的公式只需要4行就能生成，其中3行还是Python的计算公式。

这款名为handcalcs的工具包，生成LaTeX公式，但不用学LaTeX复杂的语法。平方、平方根、分式、三角函数、脚注……这个插件的功能五脏俱全，不仅能生成公式，还能计算结果。

此外，也有单位换算、格式调整等功能。

来看看它的用法。

```python
import handcalcs.render
```

```python
%%render
a = 2
b = 3
c = 2*a + b/3
```

用a、b、c表达的结果就展示了出来，而且还算出了结果。再也不用输入大段LaTeX的复杂语法，生成复杂的计算公式了。

也可以使用 `%%tex` 命令将任何Python代码段转换为有效的LaTex。例如：

```python
import handcalcs.render
from math import sqrt, pi
```

```python
%%tex
a = 2 / 3 * sqrt(pi)
```

\[
\begin{aligned}
a &= \frac{ 2 }{ 3 } \cdot \sqrt{ \pi } = \frac{ 2 }{ 3 } \cdot \sqrt{ 3.142 } &= 1.182
\end{aligned}
\]
